```python
import torch
import pandas as pd
import numpy as np

# Load the PyTorch model weights
pth_file = "../model.pth"  # Change to your file path
device = 'cpu'
state_dict = torch.load(pth_file, map_location=torch.device(device=device))

if 'state_dict' in state_dict:
    state_dict = state_dict['state_dict']

# Prepare data for DataFrame
rows = []
for name, param in state_dict.items():
    tensor = param.cpu().numpy()
    shape = list(tensor.shape)
    flattened = tensor.flatten().tolist()
    rows.append({
        "parameter": name,
        "shape": shape,
        "values": flattened
    })

# Create a DataFrame
df = pd.DataFrame(rows, columns=["parameter", "shape", "values"])

# Save to CSV
# csv_file = "model_weights_with_shapes.csv" # Save to CSV
# df.to_csv(csv_file, index=False)

# print(f"Model weights saved to {csv_file}")
```
